```python
# import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# importing CarPrice_Assignment.csv with pandas

CarData= pd.read_csv('Car_Price.csv')

CarData.head()
```

```python
# shape of the dataset
X = CarData.shape
print(X)
```

```python
# Display general information about the dataset
# This includes the count of non-null entries per column, the data type of each column, and memory usage.
CarData.info()
```

```python
# Generate descriptive statistics for numerical columns
# Provide information like mean, standard deviation, min, max, and quartiles.
CarData.describe()
```

```python
# Extract the car's company's name from the variable 'CarName' into the variable 'CarCompany'

CarData['CarCompany'] = CarData['CarName'].apply(lambda x: x.split(' ')[0])

#print(CarData['CarCompany'])
CarData.head()
```

```python
# #checking the unique values under "CarCompany"
CarData.CarCompany.unique()
```

```python
# replacing the mis-spelling with correct ones
CarData['CarCompany'] = CarData['CarCompany'].map(lambda x: str.replace(x, "porcshce", "porsche"))
CarData['CarCompany'] = CarData['CarCompany'].map(lambda x: str.replace(x, "toyouta", "toyota"))
CarData['CarCompany'] = CarData['CarCompany'].map(lambda x: str.replace(x, "vokswagen", "volkswagen"))
CarData['CarCompany'] = CarData['CarCompany'].map(lambda x: str.replace(x, "vw", "volkswagen"))
CarData['CarCompany'] = CarData['CarCompany'].map(lambda x: str.replace(x, "Nissan", "nissan"))
CarData['CarCompany'] = CarData['CarCompany'].map(lambda x: str.replace(x, "maxda", "mazda"))
# again checking the unique values
CarData.CarCompany.unique()
```

```python
#creating a dict 'company price' where key = car's company and value = median of their price

company_price = CarData.groupby('CarCompany')['price'].median().to_dict()

company_price
```

```python
# division in the buckets of low i.e. below 10000, medium i.e. range b/w 10000 and 20000 and high i.e. above 20000

def price_bucket(x):
    if x < 10000:
        return 'low'
    elif x <= 20000:
        return 'med'
    else:
        return 'high'

# mapping the company_price on the 'CarCompany' column in the dataset

CarData['CarCompany'] = CarData['CarCompany'].map(company_price).apply(price_bucket)

# checking the column's unique values

CarData.CarCompany.unique()
```

```python
#creating dummy variable for catwegorical variables
CarData.drop('CarName', axis=1, inplace=True)
categorical_cols = CarData.select_dtypes(include='object').columns
CarData = pd.get_dummies(CarData, columns=categorical_cols)

#checking the dataset
CarData.head()
```

```python
# importing required library
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

CarData_train, CarData_test = train_test_split(CarData, test_size=0.3, random_state=42)

# shape of the train and test sets
print(CarData_train.shape)
print(CarData_test.shape)
```

```python
#importing required function for scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
y_train = CarData_train['price']
X_train = CarData_train.drop('price', axis=1)

y_test = CarData_test['price']
X_test = CarData_test.drop('price', axis=1)
# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform the testing data using the same scaler
X_test_scaled = scaler.transform(X_test)
```

```python
# Train your model
from sklearn.linear_model import LinearRegression

regression = LinearRegression()
train = regression.fit(X_train_scaled, y_train)
```

```python
# Making predictions
predicted_price = regression.predict(X_test_scaled)
```

```python
# Show predictions
predicted_price
```

```python
# Plotting y_test and y_pred to understand the spread.
plt.scatter(y_test, predicted_price)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.title("y_test vs y_pred")
plt.show()
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_true=y_test, y_pred=predicted_price))
```
