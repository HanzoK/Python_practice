```python
# Read the data from the CSV file with read from pandas
data = pd.read_csv("Salary_dataset.csv")


# Configure the style
sns.set_theme(context='notebook', style='darkgrid', palette='deep', font='sans-serif', font_scale=0.8, color_codes=True, rc=None)


# Plot the data points with seaborn
#sns.scatterplot(data=data, x="YearsExperience", y="Salary")
sns.jointplot(data=data, x="YearsExperience", y="Salary", color="#8B7FB8")
```

```python
# We're going to use this library and find out how it works on the Internet
from sklearn.linear_model import LinearRegression

# Separate the features (YearsExperience) from the target variable (Salary)
X = data[['YearsExperience']]
y = data['Salary']

# Create a linear regression model
regression = LinearRegression()

# Train the model
train = regression.fit(X, y)

# Predict the Salary values based on YearsExperience
predictions = regression.predict(X)
```

```python
# Create a linear regression plot using lmplot
plot = sns.lmplot(data=data, x="YearsExperience", y="Salary", height=6, aspect=1.5)

# Set labels and title
plot.fig.suptitle("Linear Regression: Years of Experience vs Salary")
plot.fig.subplots_adjust(top=0.93)
```
